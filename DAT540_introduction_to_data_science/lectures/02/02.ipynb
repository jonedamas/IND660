- **Variables and argument passing** 
- Python supports both immutable and mutable objects
  - Immutable objects are those, whose values cannot change whereas values of mutable objects can change.
  - When assigning to a variable *"a"* in python, a reference to the object on the righthand side of the equal sign is created
  - Additionally, if the variable *"a"* is assigned to a new variable *"b"*, then *"b"* will also refer to the same object rather than maintaining a copy of the object that *"a"* refered to.
    - If *"a"* contains a immutable object (e.g: a number), and we change the value of *"a"* a new copy of the value is created and the variable *"a"* will start referring to this object
    - However, if *"a"* contains a mutable object (e.g. a list), and we change the value of *"a"* by appending a new element, a new copy will not be generated and the variables *"a"* and *"b"* will still refer to the same object

````python
# Immutable Object
a = 10
b = a
print("Original memory addresses: a=%s, b=%s" % (id(a), id(b)))
a = a + 100
print("New memory addresses: a=%s, b=%s" % (id(a), id(b)))
````

 - Example Categorizing a list of words by their first letters as a dict of lists

````python
words = ['apple', 'bat', 'bar', 'atom', 'book']

by_letter = {}

for word in words:
  letter = word[0]
  if letter not in by_letter:
    by_letter[letter] = [word]
  else:
    by_letter[letter].append(word)

print(by_letter)
````

- **Comprehensions**
 - Allows to concisely transform *List*, *Set* and *Dict* and form a new object by filtering and transforming the elements of the collection
 - **[*expr* for val in collection if condition]**
 - Also called as flattening
 - Example: if we have a for loop that does the following:
 ```python
result = []
for val in collection:
      if condition:
          result.append(expr)
```
 - It can be written as a comprehension

````python
# example
words = ['hello', 'world', 'friday', 'august', 'wednesday']

new_words = []

for i, w in enumerate(words):
    if len(w) > 5:
        new_words.append( w.title() )
    
new_words


````

````python
new_words1 = [ w.title() for w in words if len(w) > 5]
print(new_words1)
````

````python
strings = ['a', 'as', 'bat', 'car', 'dove', 'python']
[x.upper() for x in strings]
````

  - Nested Comprehensions
  - Eg.: We have multi dimentional array with names. We want to get the list of names that have atleast 2 a's

````python
all_names = [
  ['John', 'Emily', 'Michael', 'Mary', 'Steven'],
  ['Maria', 'Juan', 'Javier', 'Natalia', 'Pilar']
]
names_of_interest = []
for names in all_names:
  for name in names:
    if name.count('a') >= 2:
      names_of_interest.append(name)
print(names_of_interest)
````

 - How do we do it using *comprehensions*

````python
all_names = [
  ['John', 'Emily', 'Michael', 'Mary', 'Steven'],
  ['Maria', 'Juan', 'Javier', 'Natalia', 'Pilar']
]
names_of_interest = [name for names in all_names for name in names if name.count('a') >= 2]
print(names_of_interest)
````
